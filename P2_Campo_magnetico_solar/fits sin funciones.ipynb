```python
#MCMC 

#funcion de likelihood, sin exp
def likelihood(campo,y):
    chi_cuadrado= (1.0/2.0)*sum((campo-y)**2)
    a=(-chi_cuadrado)
    return a

#funcion 1
def modelo_lineal(t_obs,a,b):
    return (a + b*t_obs)
#funcion 2

def lineal_gauss(t_obs,c,d,sig,nu):
    return (c + d*t_obs + (1/(sig*sqrt(2*pi)))*exp(-(1/2)*((t_obs-nu)/sig)**2))
#funcion 3
def lineal_paso(t_obs, f, g, h, n, t0):
    return (f + g*t_obs + h*(1 + (2/pi)*arctan(n*(t_obs-t0))))


#listas para guardar los pasos de los parámetros libres, estan en las funciones se pueden borrar luego



f_walk = empty((0))
g_walk = empty((0))
h_walk = empty((0))
t0_walk = empty((0))
n_walk = empty((0))



like_walk3 = empty((0))

#1er append verificar randoms!! -estos se definen en las funciones, se pueden borrar luego



f_walk = np.append(f_walk,np.random.random()*200 -50)
g_walk = np.append(g_walk,np.random.random()*1 -0.5)
h_walk = np.append(h_walk,np.random.random()*200 - 100)
t0_walk = np.append(t0_walk,np.random.random()*300 +80)
n_walk = np.append(n_walk,np.random.random()*100)
```
