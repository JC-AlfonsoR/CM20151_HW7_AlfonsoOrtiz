```python
#MCMC 

#funcion de likelihood, sin exp
def likelihood(campo_m,y_m):
    chi_cuadrado= (1.0/2.0)*sum((campo_m-y_m)**2)
    a=(-chi_cuadrado)
    return a

#funcion 1
def modelo_lineal(t_obs,a,b):
    return (a + b*t_obs)
#funcion 2

def lineal_gauss(t_obs,c,d,sig,nu):
    return (c + d*t_obs + (1/(sig*sqrt(2*pi)))*exp(-(1/2)*((t_obs-nu)/sig)**2))
#funcion 3
def lineal_paso(t_obs, f, g, h, n, t0):
    return (f + g*t_obs + h*(1 + (2/pi)*arctan(n*(t_obs-t0))))


#listas para guardar los pasos de los parámetros libres, estan en las funciones se pueden borrar luego

a_walk = empty((0))
b_walk = empty((0))
c_walk = empty((0))
d_walk = empty((0))
sig_walk = empty((0))
nu_walk = empty((0))
f_walk = empty((0))
g_walk = empty((0))
h_walk = empty((0))
t0_walk = empty((0))
n_walk = empty((0))

like_walk = empty((0))
like_walk2 = empty((0))
like_walk3 = empty((0))

#1er append verificar randoms!! -estos se definen en las funciones, se pueden borrar luego

a_walk = np.append(a_walk,np.random.random()*200 -50)
b_walk = np.append(b_walk,np.random.random()*1 -0.5)
c_walk = np.append(c_walk,np.random.random()*200 -50)
d_walk = np.append(d_walk,np.random.random()*1 -0.5)
f_walk = np.append(f_walk,np.random.random()*200 -50)
g_walk = np.append(g_walk,np.random.random()*1 -0.5)
h_walk = np.append(h_walk,np.random.random()*200 - 100)
t0_walk = np.append(t0_walk,np.random.random()*300 +80)
n_walk = np.append(n_walk,np.random.random()*100)
sig_walk = np.append(sig_walk,np.random.random()+40)
nu_walk = np.append(nu_walk,np.random.random()+100)

```

```python
#modelo lineal sin funcion
#campo_=campo_m
campo_m = cols[:,1,1]

a_walk = empty((0))
b_walk = empty((0))
like_walk = empty((0))

a_walk = np.append(a_walk,np.random.random()*200 -50)
b_walk = np.append(b_walk,np.random.random()*1 -0.5)

y_i = modelo_lineal(z, a_walk[0],b_walk[0])

#areglo con la función de likelikelihood
like_walk= np.append(like_walk,likelihood(campo_m,y_i))
    
iteraciones = 2000
    
for i in range(iteraciones):
       a_prime = np.random.normal(a_walk[i], 1) 
       b_prime = np.random.normal(b_walk[i], 0.1)

       y_init = modelo_lineal(z, a_walk[i], b_walk[i])
       y_prime = modelo_lineal(z, a_prime, b_prime)
       like_prime = likelihood(campo_m, y_prime)
       like_init = likelihood(campo_m, y_init)
    
       alpha = like_prime/like_init
       if(alpha>=1.0):
            a_walk  = append(a_walk,a_prime)
            b_walk  = append(b_walk,b_prime)
            like_walk = append(like_walk, like_prime)
       else:
           beta = np.random.random()
           if(beta<=alpha):
               a_walk = append(a_walk,a_prime)
               b_walk = append(b_walk,b_prime)
               like_walk = append(like_walk, like_prime)
           else:
               a_walk = append(a_walk,a_walk[i])
               b_walk = append(b_walk,b_walk[i])
               like_walk = append(like_walk, like_init)

print a_walk, b_walk, like_walk
print len(a_walk), len(b_walk), len(like_walk) 
```
